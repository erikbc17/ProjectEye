```python
'''Importing the Libraries'''MSGSzZxsxsx
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.metrics import metrics
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input, Activation, BatchNormalization
from keras import optimizers
from keras import regularizers
from keras.preprocessing.image import ImageDataGenerator
import os
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
'''Importing the Data'''
drive.mount('/content/drive')



datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input)

path_train_data = '/content/drive/Shareddrives/Machine Learning/projectguns/train'
X_train = datagen.flow_from_directory(directory=path_train_data, target_size=(50,50),classes=['gun','not_gun'], batch_size=10)

path_valid_data = '/content/drive/Shareddrives/Machine Learning/projectguns/valid'
X_valid = datagen.flow_from_directory(directory=path_valid_data, target_size=(50,50),classes=['gun','not_gun'], batch_size=10)

path_test_data = '/content/drive/Shareddrives/Machine Learning/projectguns/test'
X_test = datagen.flow_from_directory(directory=path_test_data, target_size=(50,50),classes=['gun','not_gun'], batch_size=10)
```

```python
''' Machine Learning Model'''

# Creating the CNN Model
model = Sequential(name='gun_CNN_classifier')

# Set of Layers
model.add(Input(shape=(50,50,3), name='input'))
model.add(Flatten(name='flatten'))
model.add(Dense(32, activation='relu', kernel_initializer='random_normal', kernel_regularizer=regularizers.L2(1e-3),name='dense1'))
model.add(Dense(64, activation='relu', kernel_regularizer=regularizers.L2(1e-3), name='dense2'))
model.add(Dense(128, activation='relu', kernel_regularizer=regularizers.L2(1e-3), name='dense3'))
model.add(Dense(256, activation='relu', kernel_regularizer=regularizers.L2(1e-3), name='dense4'))
model.add(Dense(2, activation='softmax', name='Output'))

# Compilation of Model

model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=["accuracy"])

model.summary()

# fitting the model
model.fit(x=X_train, validation_data=X_valid, epochs=50, verbose=1)
scores = model.evaluate(x=X_valid)
print('Accuracy:',+scores[1]*100)
```

```python
# Prediction

y_train_pred = model.predict(x=X_train)
y_test_pred = model.predict(x=X_test)
y_valid_pred = model.predict(x=X_valid)
```
